# How safe is it to ride a bike in NYC?

```python
# Import Dependencies
import pandas as pd
import calendar
import datetime
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = "data/NYPD_Motor_Vehicle_Collisions_dev_data.csv"
eclipse_df = pd.read_csv(csv_path)
eclipse_df.head()
```

```python
# Get a reference to the column names
eclipse_df.columns
```

```python
# Drop columns
eclipse_df = eclipse_df.drop(['LOCATION', 'CROSS STREET NAME', 'OFF STREET NAME',
             'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4','NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
       'CONTRIBUTING FACTOR VEHICLE 5', 'UNIQUE KEY', 'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED',
       'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4',
       'VEHICLE TYPE CODE 5'], axis=1)
eclipse_df.columns
```

```python
#total number of accidents
accidents_total = len(eclipse_df)
accidents_total
```

```python
#drop all rows not invilving ped or cycl injuries
#cycling_injuries_df = eclipse_df.loc[(eclipse_df ['NUMBER OF CYCLIST INJURED'] > 0) | (eclipse_df ['NUMBER OF CYCLIST KILLED']>0)]
eclipse_df.drop(eclipse_df[(eclipse_df ['NUMBER OF CYCLIST INJURED'] <=0 )
                              & (eclipse_df ['NUMBER OF CYCLIST KILLED'] <= 0 )
                           & (eclipse_df ['NUMBER OF PEDESTRIANS INJURED'] <= 0 )
                              & (eclipse_df ['NUMBER OF PEDESTRIANS KILLED'] <= 0)].index,inplace=True)
len(eclipse_df)
```

```python
#Extract 'Month' from 'DATE' for aggregation
eclipse_df['DATE'] = pd.to_datetime(eclipse_df['DATE'])
eclipse_df['MONTH'] = pd.DatetimeIndex(eclipse_df['DATE']).month
eclipse_df['YEAR'] = pd.DatetimeIndex(eclipse_df['DATE']).year
eclipse_df.head()
```

```python
# create aggregate monthly stats 
#Finding the ratios of % of cyclist killed over total number of persons killed,  % of pedestrians killed over total numeber of persons killes, 
# % of cyclist injured over total number of persons injured, % of pedestrians injured over total numeber of persons injured
#number  of accidents involving cyclists
col_names =  ['CYCLIST_KILLED', 'CYCLIST_INJURED', 'PEDESTRIANS KILLED', 'PEDESTRIANS INJURED', 'ACCIDENTS', 'DATE']
my_df  = pd.DataFrame(columns = col_names)
my_df['CYCLIST_INJURED'] = eclipse_df.groupby(['YEAR', 'MONTH'])['NUMBER OF CYCLIST INJURED'].sum()
my_df['CYCLIST_KILLED'] = eclipse_df.groupby(['YEAR', 'MONTH'])['NUMBER OF CYCLIST KILLED'].sum()
my_df['PEDESTRIANS KILLED'] = eclipse_df.groupby(['YEAR', 'MONTH'])['NUMBER OF PEDESTRIANS KILLED'].sum()
my_df['PEDESTRIANS INJURED'] = eclipse_df.groupby(['YEAR', 'MONTH'])['NUMBER OF PEDESTRIANS INJURED'].sum()
#my_df['CYCLIST INVOLVED'] = eclipse_df.groupby(['YEAR', 'MONTH'])(['VEHICLE TYPE CODE 1']=='Bike').count()
my_df['ACCIDENTS'] = eclipse_df.groupby(['YEAR', 'MONTH']).count()
#my_df['DATE'] = datetime.date(year = my_df.iloc(0), month= my_df.iloc(1), day = 1)

my_df
```

```python
my_df.reset_index(inplace = True)
my_df
```

```python
for index, row in my_df.iterrows():
    year = row['YEAR']
    month= row['MONTH']
    my_df.at[index, 'DATE']= datetime.date(year , month, day = 1)
my_df
```

```python
# result csv for analysis quesiton #1
my_df.to_csv('output/accident_data.csv', encoding='utf-8', index=False)
```
